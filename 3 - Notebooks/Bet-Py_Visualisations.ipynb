```python
## Importation des modules
import os

import numpy             as np
import pandas            as pd
import seaborn           as sns
import matplotlib.pyplot as plt
```

```python
## Chargement des données
os.chdir(r'D:\Users\User\Documents\Works\03 - Formations\02 - DataScientest\06 - Project\00 - Data\Football')

match_results = pd.read_csv('match_results.csv', index_col=0)
indics        = pd.read_csv('indics.csv',        index_col=0)
```

```python
## Quelques nuages de points
fig = plt.figure(figsize=(15, 10))
ax  = fig.add_subplot(1, 1, 1)

ax.scatter(indics[indics.Resultat == 1.]['Ecart_diff'], indics[indics.Resultat == 1.]['Ecart_pour'], marker='+', s=9., c='red',  label='Victoire')
ax.scatter(indics[indics.Resultat == 0.]['Ecart_diff'], indics[indics.Resultat == 0.]['Ecart_pour'], marker='+', s=9., c='blue', label='Défaite')

ax.set_xlabel('Écart entre les différences de buts')
ax.set_ylabel('Écart entre les nombres de buts marqués')
ax.legend(loc='upper left');
```

```python
## Quelques nuages de points
fig = plt.figure(figsize=(15, 10))
ax  = fig.add_subplot(1, 1, 1)

ax.scatter(indics[indics.Resultat == 1.]['Ecart_diff'], indics[indics.Resultat == 1.]['Ecart_contre'], marker='+', s=9., c='red',  label='Victoire')
ax.scatter(indics[indics.Resultat == 0.]['Ecart_diff'], indics[indics.Resultat == 0.]['Ecart_contre'], marker='+', s=9., c='blue', label='Défaite')

ax.set_xlabel('Écart entre les différences de buts')
ax.set_ylabel('Écart entre les nombres de buts encaissés')
ax.legend(loc='upper left');
```
