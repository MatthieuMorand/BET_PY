```python
## Création d'une fonction pour constituer la table « match_results »
def collect_results(df, k=5):
    """
    Cette fonction permet de créé la table « match_results » (en cours d'élaboration)
    à partir de la table « match ».
    
    Arguments :
     - df : la table « match » ;
     - k  : représente le nombre de match passés à regarder pour la « streak » de victoires (par défaut 5).
    
    Prérequis : « df » doit être trié sur la colonne « dateutc » (ordre croissant).
    """
    dico_streak        = {}
    dico_goals_for     = {}
    dico_goals_against = {}
    
    # Période de la journée
    #  - 0 : [10 h - 12 h[
    #  - 1 : [12 h - 14 h[
    #  - 2 : [14 h - 16 h[
    #  - 3 : [16 h - 18 h[
    #  - 4 : [18 h - 20 h[
    #  - 5 : [20 h - 22 h[
    #  - 6 : [22 h - 24 h[
    periods = {}
    for i in range(10, 21, 2):
        val = i / 2 - 5
        periods[i    ] = val
        periods[i + 1] = val
    
    ## Composition de la table « match_results »
    dico_res = {}
    
    # Variables sur le match
    dico_res['Date']                = [] # - Date et heure du match
    dico_res['Club']                = [] # - Nom du club
    dico_res['Club_ADV']            = [] # - Nom du club adverse
    dico_res['Dom_Ext']             = [] # - Booléen (1 ou 0) indiquant si le club joue à domicile ou à l'extérieur
    dico_res['Periode_jour']        = [] # - Période de la journée dans laquelle se déroule le match
    dico_res['Arbitre_id']          = [] # - Identifiant de l'arbitre du match
    dico_res['Match_id']            = [] # - Identifiant du match
    
    # Variables sur le club
    dico_res['CLB_Serie_vic']       = [] # - Nombre de matchs remportés par le club (sur les k derniers matchs)
    dico_res['CLB_Tot_buts_pour']   = [] # - Nombre de buts marqués dans la saison
    dico_res['CLB_Tot_buts_contre'] = [] # - Nombre de buts encaissés dans la saison
    dico_res['CLB_Tot_buts_diff']   = [] # - Différence de buts sur la saison
    
    # Variables sur le club adverse
    dico_res['ADV_Serie_vic']       = [] # - Nombre de matchs remportés par le club (sur les k derniers matchs)
    dico_res['ADV_Tot_buts_pour']   = [] # - Nombre de buts marqués dans la saison
    dico_res['ADV_Tot_buts_contre'] = [] # - Nombre de buts encaissés dans la saison
    dico_res['ADV_Tot_buts_diff']   = [] # - Différence de buts sur la saison
    
    # Variables à prédire
    dico_res['[Resultat]']          = [] # - Valeur (1, 0.5 ou 0) indiquant si le club à gagné, fait nul, ou perdu le match
    dico_res['[Buts_dom]']          = [] # - Nombre de buts marqués par l'équipe jouant à domicile
    dico_res['[Buts_ext]']          = [] # - Nombre de buts marqués par l'équipe jouant à l'extérieur
    
    for _, row in df.iterrows():
        # Calcul de la valeur correspondant au résultat du match
        vic_home = 1. if row.goal_by_home_club > row.goal_by_away_club else 0.
        
        # Gestion de la série de victoires
        if row.home_club not in dico_streak.keys():
            dico_streak[row.home_club] = [vic_home]
        else:
            dico_streak[row.home_club].append(vic_home)
        
        if row.away_club not in dico_streak.keys():
            dico_streak[row.away_club] = [1. - vic_home]
        else:
            dico_streak[row.away_club].append(1. - vic_home)
        
        # Gestion des buts pour
        if row.home_club not in dico_goals_for.keys():
            dico_goals_for[row.home_club] = (0., row.goal_by_home_club)
        else:
            dico_goals_for[row.home_club] = (sum(dico_goals_for[row.home_club]), row.goal_by_home_club)
        
        if row.away_club not in dico_goals_for.keys():
            dico_goals_for[row.away_club] = (0., row.goal_by_away_club)
        else:
            dico_goals_for[row.away_club] = (sum(dico_goals_for[row.away_club]), row.goal_by_away_club)
        
        # Gestion des buts contre
        if row.home_club not in dico_goals_against.keys():
            dico_goals_against[row.home_club] = (0., row.goal_by_away_club)
        else:
            dico_goals_against[row.home_club] = (sum(dico_goals_against[row.home_club]), row.goal_by_away_club)
        
        if row.away_club not in dico_goals_against.keys():
            dico_goals_against[row.away_club] = (0., row.goal_by_home_club)
        else:
            dico_goals_against[row.away_club] = (sum(dico_goals_against[row.away_club]), row.goal_by_home_club)
        
        # Ajout du premier club (celui jouant à domicile)
        dico_res['Date'].append(row.dateutc)
        dico_res['Club'].append(row.home_club)
        dico_res['Club_ADV'].append(row.away_club)
        dico_res['Dom_Ext'].append(1.)
        dico_res['Periode_jour'].append(periods[row.dateutc.hour])
        dico_res['Arbitre_id'].append(row.referee_id)
        dico_res['Match_id'].append(row.id)
        
        dico_res['CLB_Serie_vic'].append(sum(dico_streak[row.home_club][-min(k+1, len(dico_streak[row.home_club])):-1]))
        buts_pour   = dico_goals_for[row.home_club][0]
        buts_contre = dico_goals_against[row.home_club][0]
        dico_res['CLB_Tot_buts_pour'].append(buts_pour)
        dico_res['CLB_Tot_buts_contre'].append(buts_contre)
        dico_res['CLB_Tot_buts_diff'].append(buts_pour - buts_contre)
        
        dico_res['ADV_Serie_vic'].append(sum(dico_streak[row.away_club][-min(k+1, len(dico_streak[row.away_club])):-1]))
        buts_pour   = dico_goals_for[row.away_club][0]
        buts_contre = dico_goals_against[row.away_club][0]
        dico_res['ADV_Tot_buts_pour'].append(buts_pour)
        dico_res['ADV_Tot_buts_contre'].append(buts_contre)
        dico_res['ADV_Tot_buts_diff'].append(buts_pour - buts_contre)
        
        dico_res['[Resultat]'].append(vic_home)
        dico_res['[Buts_dom]'].append(row.goal_by_home_club)
        dico_res['[Buts_ext]'].append(row.goal_by_away_club)
        
        # Ajout du second club (celui jouant à l'extérieur)
        dico_res['Date'].append(row.dateutc)
        dico_res['Club'].append(row.away_club)
        dico_res['Club_ADV'].append(row.home_club)
        dico_res['Dom_Ext'].append(0.)
        dico_res['Periode_jour'].append(periods[row.dateutc.hour])
        dico_res['Arbitre_id'].append(row.referee_id)
        dico_res['Match_id'].append(row.id)
        
        dico_res['CLB_Serie_vic'].append(sum(dico_streak[row.away_club][-min(k+1, len(dico_streak[row.away_club])):-1]))
        buts_pour   = dico_goals_for[row.away_club][0]
        buts_contre = dico_goals_against[row.away_club][0]
        dico_res['CLB_Tot_buts_pour'].append(buts_pour)
        dico_res['CLB_Tot_buts_contre'].append(buts_contre)
        dico_res['CLB_Tot_buts_diff'].append(buts_pour - buts_contre)
        
        dico_res['ADV_Serie_vic'].append(sum(dico_streak[row.home_club][-min(k+1, len(dico_streak[row.home_club])):-1]))
        buts_pour   = dico_goals_for[row.home_club][0]
        buts_contre = dico_goals_against[row.home_club][0]
        dico_res['ADV_Tot_buts_pour'].append(buts_pour)
        dico_res['ADV_Tot_buts_contre'].append(buts_contre)
        dico_res['ADV_Tot_buts_diff'].append(buts_pour - buts_contre)
        
        dico_res['[Resultat]'].append(1. - vic_home)
        dico_res['[Buts_dom]'].append(row.goal_by_home_club)
        dico_res['[Buts_ext]'].append(row.goal_by_away_club)
        
    return pd.DataFrame(data=dico_res)
```

```python
## Création de la table « match_results » à partir de la table « match »
match_results = collect_results(match.sort_values(by=['dateutc']))

# Écriture de la table « match_results »
match_results.to_csv('match_results.csv', sep=',')

# Affichage
print('Cardinalités de la table : {} x {}'.format(match_results.shape[0], match_results.shape[1]))
match_results.tail(10)
```

```python
## Études sur les relations entre les variables de la table « match_results »
plt.figure(figsize=(13, 13))
sns.heatmap(match_results.corr(), annot=True, cmap='RdBu_r', center=0);

## Quelques résultats

## Variables ayant un influence sur la variable cible (« [Resultat] ») :
## - Tot_buts_diff (influence médiocre (27 %))
## - Dom_Ext (influence médiocre (21 %)) (variable indépendante)
## - Serie_vic (influence médiocre (21 %))
## - Tot_buts_pour (influence médiocre (18 %))
```

```python
## Importance de la variable « Dom_Ext » sur la variable « [Resultat] »
cm = pd.crosstab(match_results['[Resultat]'], match_results.Dom_Ext, rownames=['Résultat'], colnames=['Lieu du match'])

print(cm)
print('-> {} % des matchs ont été gagnés à domicile.'.format(round(cm[1][1] / (cm[1][0] + cm[1][1]) * 100., 2)))
```

```python
## Importance des variables « CLB_Tot_buts_diff » et « ADV_Tot_buts_diff » sur la variable « [Resultat] »
df_diff = pd.DataFrame(match_results.CLB_Tot_buts_diff > match_results.ADV_Tot_buts_diff).replace({True: 1., False: 0.}).rename(columns={0: 'Diff'})

cm = pd.crosstab(match_results['[Resultat]'], df_diff.Diff, rownames=['Résultat'], colnames=['Meilleur différence'])

print(cm)
print("-> {} % des équipes ayant perdus un match avait une moindre différence de buts que l'équipe adverse sur la saison.".format(round(cm[0][0] / (cm[0][0] + cm[0][1]) * 100, 2)))
print("-> {} % des équipes ayant gagnés un match avait une meilleure différence de buts que l'équipe adverse sur la saison.".format(round(cm[1][1] / (cm[1][0] + cm[1][1]) * 100, 2)))
```

```python
## Importance des variables « CLB_Tot_buts_pour » et « ADV_Tot_buts_pour » sur la variable « [Resultat] »
df_diff = pd.DataFrame(match_results.CLB_Tot_buts_pour > match_results.ADV_Tot_buts_pour).replace({True: 1., False: 0.}).rename(columns={0: 'Diff'})

cm = pd.crosstab(match_results['[Resultat]'], df_diff.Diff, rownames=['Résultat'], colnames=['Meilleur différence'])

print(cm)
print("-> {} % des équipes ayant perdus un match avait marquées moins de buts que l'équipe adverse sur la saison.".format(round(cm[0][0] / (cm[0][0] + cm[0][1]) * 100, 2)))
print("-> {} % des équipes ayant gagnés un match avait marquées plus de buts que l'équipe adverse sur la saison.".format(round(cm[1][1] / (cm[1][0] + cm[1][1]) * 100, 2)))
```

```python
## Importance des variables « CLB_Tot_buts_contre » et « ADV_Tot_buts_contre » sur la variable « [Resultat] »
df_diff = pd.DataFrame(match_results.CLB_Tot_buts_contre < match_results.ADV_Tot_buts_contre).replace({True: 1., False: 0.}).rename(columns={0: 'Diff'})

cm = pd.crosstab(match_results['[Resultat]'], df_diff.Diff, rownames=['Résultat'], colnames=['Meilleur différence'])

print(cm)
print("-> {} % des équipes ayant perdus un match avait encaissés plus de buts que l'équipe adverse sur la saison.".format(round(cm[0][0] / (cm[0][0] + cm[0][1]) * 100, 2)))
print("-> {} % des équipes ayant gagnés un match avait encaissés moins de buts que l'équipe adverse sur la saison.".format(round(cm[1][1] / (cm[1][0] + cm[1][1]) * 100, 2)))
```
